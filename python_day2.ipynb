# Plotting data with plotnine

```python
surveys_complete = surveys_df.dropna()
```

```python
surveys_complete
```

```python
(p9.ggplot(data=surveys_complete))
```

```python
(p9.ggplot(data=surveys_complete,
           mapping=p9.aes(x='weight', y='hindfoot_length')
          ))
```

```python
(p9.ggplot(data=surveys_complete,
           mapping=p9.aes(x='weight', y='hindfoot_length'))
           + p9.geom_point()
          )
```

```python
(p9.ggplot(data=surveys_complete,
           mapping=p9.aes(x='weight', y='hindfoot_length'))
           + p9.geom_point(color='blue', alpha=0.1)
          )
```

```python
(p9.ggplot(data=surveys_complete,
           mapping=p9.aes(x='weight', y='hindfoot_length', color='species_id'))
           + p9.geom_point(alpha=0.1)
          )
```

```python
(p9.ggplot(data=surveys_complete,
           mapping=p9.aes(x='weight', y='hindfoot_length', color='species_id'))
           + p9.geom_point(alpha=0.1)
           + p9.xlab("Weight (g)")  
          )
```

```python
(p9.ggplot(data=surveys_complete,
           mapping=p9.aes(x='weight', y='hindfoot_length', color='species_id'))
           + p9.geom_point(alpha=0.1)
           + p9.xlab("Weight (g)")  
           + p9.ylab("Hindfoot length")
          )
```

```python
(p9.ggplot(data=surveys_complete,
           mapping=p9.aes(x='weight', y='hindfoot_length', color='species_id'))
           + p9.geom_point(alpha=0.1)
           + p9.xlab("Weight (g)")  
           + p9.ylab("Hindfoot length")
           + p9.ggtitle("Hindfoot length by weight")
          )
```

```python
hindfoot_weight_plot = (p9.ggplot(data=surveys_complete,
           mapping=p9.aes(x='weight', y='hindfoot_length', color='species_id'))
           + p9.geom_point(alpha=0.1)
           + p9.xlab("Weight (g)")  
           + p9.ylab("Hindfoot length")
           + p9.ggtitle("Hindfoot length by weight")
          )
```

```python
hindfoot_weight_plot.save("hfw_plot.png", width=5, height=5, dpi=300)
```

```python
(p9.ggplot(data=surveys_complete,
           mapping=p9.aes(x='weight', y='hindfoot_length', color='species_id'))
           + p9.geom_point(alpha=0.1)
           + p9.xlab("Weight (g)")  
           + p9.ylab("Hindfoot length")
           + p9.ggtitle("Hindfoot length by weight")
           + p9.facet_wrap("sex")
          )
```

```python
(p9.ggplot(data=surveys_complete,
           mapping=p9.aes(x='weight', y='hindfoot_length', color='species_id'))
           + p9.geom_point(alpha=0.1)
           + p9.xlab("Weight (g)")  
           + p9.ylab("Hindfoot length")
           + p9.ggtitle("Hindfoot length by weight")
           + p9.facet_wrap("year")
          )
```

```python
(p9.ggplot(data=surveys_complete,
            mapping=p9.aes(x='factor(species_id)'))
            + p9.geom_bar()
            + p9.ggtitle("Species counts") 
)
```
